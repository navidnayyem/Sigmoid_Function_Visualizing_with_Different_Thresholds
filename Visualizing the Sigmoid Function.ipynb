```python
import numpy as np # Import numpy library for numerical operations
import math # Import math library for mathematical functions
import matplotlib.pyplot as plt # Import matplotlib library for plotting
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x)) # Define the sigmoid function using numpy for efficient computation
```

```python
# Test the sigmoid function with different values
test_values = np.linspace(-10, 10, 100) # Generate 100 evenly spaced values from -10 to 10
sigmoid_values = sigmoid(test_values) # Calculate sigmoid values for the test values
```

```python
# Visualize the sigmoid function
plt.figure(figsize=(8, 6)) # Create a figure with a specific size
plt.plot(test_values, sigmoid_values, label='Sigmoid Function', color='b') # Plot the sigmoid curve
plt.title('Sigmoid Function') # Set the title of the plot
plt.xlabel('x') # Set the label for the x-axis
plt.ylabel('sigmoid(x)') # Set the label for the y-axis
plt.axhline(y=0.5, color='r', linestyle='--', label='y=0.5') # Add a horizontal line at y=0.5 for reference
plt.grid(True) # Add gridlines to the plot
plt.legend() # Display the legend
plt.show() # Show the plot
```
